```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
data_loc = '/Users/simon/Documents/DTU/9. semester/deep learning/data'

measurements_all = [pd.read_csv(os.path.join(data_loc, f'Case{case}',f'Proj_Measurements_Case{case}.csv'), parse_dates=['Date_Time']) for case in (1,2,3)]
nwp_all = [pd.read_csv(os.path.join(data_loc, f'Case{case}',f'Proj_NWP_Case{case}.csv'), parse_dates=['Date_Time']) for case in (1,2,3)]

for i in range(3):
    measurements_all[i].name = f'Measurement {i+1}'
    nwp_all[i].name = f'NWP {i+1}'

df_all = measurements_all+nwp_all
    
```

```python
for df in df_all:
    print('Dataframe: ',df.name)
    print('\nVariable names:    Number of nan:')
    print(df.isna().sum())

    print(f'\nNumber of observations: {len(df.index)}. Memory usage: {df.memory_usage().sum()/1048576:.2f} mb')

    display(df.describe())


    print('Minimum time difference:', df['Date_Time'].diff().min())
    print('Maximum time difference:', df['Date_Time'].diff().max())
```

NWP has no missing values, but the time points are not uniformly distributed.


Measurement has missing values for all variables except for the date_time, so we know that the time stamps are all evenly distributed with 15 minutes steps.
